## Advanced Data Aggregation

```python
df = pd.read_csv(r"C:\Users\Owner\Desktop\IMP\udm\Project_06_Materials\Project_07_Materials\Project_07_Materials\us_baby_names.csv")
```

```python
df
```

```python
df.info()
```

```python
df.groupby(["Name","Gender"]).Count.sum().reset_index(1)
```

```python
agg = df.groupby(["Name","Gender"]).agg(Total =("Count","sum"),
                                        Duration =("Name","count"),
                                         First_yr =("Year","min"),
                                          Last_yr = ("Year","max"),
                                           Max_count =("Count","max"))
```

```python
agg
```

```python
def best_year(group):
    return group.nlargest(1,"Count").Year
```

```python
best  = df.groupby(["Name","Gender"]).apply(best_year)
```

```python
best
```

```python
best.droplevel(-1)
```

```python
agg["Best_year"] = best.droplevel(-1)
```

```python
 agg
```

```python
 agg 
```

```python
 agg.reset_index(drop=True)
```

```python
agg[(agg["Name"]=="Aabir") & (agg["Gender"]=="M")]
```
